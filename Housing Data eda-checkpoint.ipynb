```python
# Scatter plots for MEDV vs other numerical features
for col in data.columns:
    if col != 'MEDV' and data[col].dtype in ['float64', 'int64']:
        plt.figure(figsize=(8, 5))
        sns.scatterplot(x=data[col], y=data['MEDV'], edgecolor='w')
        plt.title(f"Scatter plot of {col} vs MEDV")
        plt.xlabel(col)
        plt.ylabel("MEDV")
        plt.show()
```

```python
# Missing values visualization
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr().tail(1), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Histograms for numerical features
data.hist(bins=15, figsize=(20, 15), layout=(4, 4), edgecolor="black")
plt.tight_layout()
plt.show()
```

```python
# Boxplots for outlier detection
plt.figure(figsize=(20, 10))
sns.boxplot(data=data, orient="h", palette="Set2")
plt.title("Boxplots for Outlier Detection")
plt.show()
```
